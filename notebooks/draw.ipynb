```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display
# plt.rcParams['font.sans-serif'] = 'SimHei'
colors = [(102, 0, 32),(177, 24, 45),(212, 97, 83),(242, 165, 132),(253, 219, 196),(145, 197, 218),(71, 146, 196),(36, 100, 171)]
colors = [(r / 255, g / 255, b / 255) for r, g, b in colors]
fig, ax = plt.subplots()
color_matrix = np.array([colors])
ax.imshow(color_matrix, aspect='auto', extent=(0, len(colors), 0, 1))
ax.set_xlabel('Color Index')
ax.set_title('Color Palette')
ax.set_yticks([])
ax.set_yticklabels([])
ax.set_xticks(np.arange(len(colors)) + 0.5)
ax.set_xticklabels(range(len(colors)))
plt.show()
%matplotlib inline
```

```python
labels = ['C', 'N', 'O', 'S', 'F', 'Si', 'P', 'Cl', 'Br','I']
values = [96, 55, 90, 10, 13, 4, 4, 4, 1, 0]

# 创建柱状图
plt.bar(labels, values, color = colors[6])

for i, v in enumerate(values):
    plt.text(i, v + 2, str(v), ha='center', va='bottom', fontsize=10, color='black')

# 添加标签和标题
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count of different elements in homopolymer test dataset')

# 显示图表
plt.show()
```

```python
def DrawSMILES(smiles):
    print("SMILES:", smiles)
    mol = Chem.MolFromSmiles(smiles)
    img = Draw.MolToImage(mol)
    display(img)
```

```python
DrawSMILES("[*]Oc1ccc(-c2ccc(C([*])=O)cc2)cc1")
```

```python
DrawSMILES("[*:1]c1cc(F)c([*:2])cc1F.[*:3]c1cc(Cl)c([*:4])cc1C")
```
